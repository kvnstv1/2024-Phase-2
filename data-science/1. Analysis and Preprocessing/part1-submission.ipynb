```python
import sklearn
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

## 1. Find all variables and understand them

```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

from sklearn.preprocessing import LabelEncoder, OneHotEncoder

from pandas import read_csv

#1.1 Load chosen datasets from the datasets folder into your notebook and display the first ten instances.

df = pd.read_csv("store_sales.csv", encoding = 'latin-1')
df.loc[0:9]

print("")

print("The data types of each column are as follows:")

df.dtypes

#1.2 Provide key statistical measures such as mean and standard deviation without hardcoding them.

mean_sales = df['Sales'].mean()
mean_profit = df['Profit'].mean()
std_sales = df['Sales'].std()
std_profit = df['Profit'].std()

print(f"The mean of the sales is ${mean_sales:.2f} and the mean of the profit is ${mean_profit:.2f}")
print(f"The standard deviation of the sales is ${std_sales:.2f} and that of the profit is ${std_profit:.2f}")



#1.3 Visualize numerical columns through appropriate plots, e.g. bar plots, histograms, box and whisker plots, etc.

plt.figure(figsize=(8,6))
plt.bar(df['Sub-Category'], df['Sales'])
plt.title('Items sold vs. sales')
plt.xlabel('Sub-categories')
plt.ylabel('Sales')
plt.show()


plt.figure(figsize=(8,6))
plt.hist(df['Sub-Category'], bins=30, edgecolor='black')
plt.title('Histogram')
plt.xlabel('Sub-Category')
plt.ylabel('Frequency')
plt.show()


plt.figure(figsize=(8,6))
plt.boxplot(df['Discount'], labels=['Discount'])
plt.title('Box and whisker plot of the discount')
plt.ylabel('Discount')
plt.show()



#1.4 Convert data types of columns as needed, ensuring all values fed into machine learning models are numerical.

label_encoder = LabelEncoder()
df['Sub-Category'] = label_encoder.fit_transform(df['Sub-Category'])
#print(df['Sub-Category'])

df['Ship Mode'] = label_encoder.fit_transform(df['Ship Mode'])
#print(df['Ship Mode'])

df['Region'] = label_encoder.fit_transform(df['Region'])
#print(df['Region'])

df['City'] = label_encoder.fit_transform(df['City'])
#print(df['City'])
print(df['City'].nunique)




```

## 2. Clean data

```python
# Missing values 

missing_values = df.isnull()
print(missing_values.sum())
print("wow no missing data?")
print("")

na_values = df.isna()
print(na_values.sum())

#Outliers

def find_outliers_IQR(df):
    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR)))]
    return outliers

outliers_sales = find_outliers_IQR(df['Sales'])
print("Number of outliers:", len(outliers_sales))
print("Max outlier value:", outliers_sales.max())
print("Min outlier value:", outliers_sales.min())


outliers_profit = find_outliers_IQR(df['Profit'])
print("Number of outliers:", len(outliers_profit))
print("Max outlier value:", outliers_profit.max())
print("Min outlier value:", outliers_profit.min())

#Imbalanced data



#Feature selection

df = df.drop('Order ID', axis=1)
df = df.drop('Customer ID', axis=1)
df = df.drop('Product ID', axis=1)
df = df.drop('Product Name', axis=1)
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
df['Order Date'] = pd.to_datetime(df['Order Date'])


```

## 3. Visualise data

```python
# Identify trends/patterns using non-numerical categories

plt.figure(figsize=(8,6))
plt.bar(df['City'], df['Sales'])
plt.title('City vs. sales')
plt.xlabel('City')
plt.ylabel('Sales')
plt.show()
```
